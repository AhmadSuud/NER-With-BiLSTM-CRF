```python
import pandas as pd
```

```python
datas =pd.read_csv('dfbersih6-7.csv')
```

```python
datas
```

```python
dataku = pd.DataFrame(columns=['Artikel', 'Rangkuman'])
```

```python
dataku
```

```python
for data in datas.values:
  artikel = [data[0]]
  Rangkuman = data[1].split()
  artikel[len (artikel):] = [None]*(len (Rangkuman)-1)
  dt = pd.DataFrame({'Artikel': artikel, 'Rangkuman': Rangkuman})
  dataku = pd.concat([dataku, dt])
```

```python
dataku
```

```python
dataku.to_csv('fix4.csv', index=False)
```
